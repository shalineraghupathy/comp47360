```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.neural_network import MLPRegressor
from sklearn.metrics import mean_squared_error, r2_score
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
```

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
df = pd.read_csv('parks_busyness_weather.csv')
```

```python
df.head()
```

```python
# Separate features and target
X = df.drop(columns='target')
y = df['target']
```

```python
training_set,testing_set=train_test_split(df,test_size=0.3, random_state=12)
```

```python
weekend_2D = training_set['weekend'].values.reshape(-1, 1)
print(weekend_2D)
encoder_weekend=OneHotEncoder()
encoder_weekend.fit(weekend_2D)
encoded_weekend = encoder_weekend.transform(weekend_2D)
```

```python
season_2D = training_set['season'].values.reshape(-1, 1)
print(season_2D)
encoder_season=OneHotEncoder()
encoder_season.fit(season_2D)
encoded_season = encoder_season.transform(season_2D)
```

```python
holiday_2D = training_set['isHoliday?'].values.reshape(-1, 1)
print(holiday_2D)
encoder_holiday=OneHotEncoder()
encoder_holiday.fit(holiday_2D)
encoded_holiday = encoder_holiday.transform(holiday_2D)
```

```python

trainset_rf_encoded_weekend_df = pd.DataFrame(encoded_weekend.toarray(), columns=encoder_weekend.get_feature_names_out(['weekend']))
trainset_rf_encoded_season_df=pd.DataFrame(encoded_season.toarray(),columns=encoder_season.get_feature_names_out(['season']))
trainset_rf_encoded_holiday_df=pd.DataFrame(encoded_holiday.toarray(),columns=encoder_holiday.get_feature_names_out(['isHoliday?']))
#trainset_rf_encoded_month_df=pd.DataFrame(encoded_month.toarray(),columns=encoder_month.get_feature_names_out(['month']))
#trainset_rf_encoded_day_df=pd.DataFrame(encoded_day.toarray(),columns=encoder_day.get_feature_names_out(['day']))
#trainset_rf_encoded_hour_df=pd.DataFrame(encoded_hour.toarray(),columns=encoder_hour.get_feature_names_out(['hour']))
```

```python

training_set.reset_index(drop=True, inplace=True)
trainset_rf_encoded_weekend_df.reset_index(drop=True, inplace=True)
trainset_rf_encoded_season_df.reset_index(drop=True, inplace=True)
trainset_rf_encoded_holiday_df.reset_index(drop=True, inplace=True)
#trainset_rf_encoded_month_df.reset_index(drop=True, inplace=True)
#trainset_rf_encoded_day_df.reset_index(drop=True, inplace=True)
#trainset_rf_encoded_hour_df.reset_index(drop=True, inplace=True)

```

```python
training_set
```

```python
training_set= pd.concat([training_set, trainset_rf_encoded_weekend_df, trainset_rf_encoded_season_df,trainset_rf_encoded_holiday_df], axis=1)
```

```python
mlp_features= [ 'weekend_0', 'weekend_1', 'season_1', 'season_2', 'season_3',
       'season_4', 'isHoliday?_0', 'isHoliday?_1','rain_1h','rain_3h']


x_train=training_set[mlp_features]
y_train=training_set['park_busyness']

print("\nDescriptive features in X:\n", x_train)
print("\nTarget feature in y:\n", y_train)
```

```python
# Apply PCA to reduce the dimensionality
pca = PCA(n_components=0.95)  # Retain 95% of variance
X_pca = pca.fit_transform(x_train)
```

```python
# Initialize the MLP Regressor
mlp = MLPRegressor(hidden_layer_sizes=(100, 50,25), max_iter=1000, random_state=12)

# Train the model
mlp.fit(x_train, y_train)
```

```python
# Make predictions on the test set
prediction = mlp.predict(x_train)
```

```python
prediction
```

```python
# Predicted class labels for all examples, 
# using the trained model, on in-sample data (same sample used for training and test)
mlp_predictions = mlp.predict(x_train)
df_true_vs_rfr_predicted = pd.DataFrame({'ActualClass': y_train, 'PredictedClass': mlp_predictions})
df_true_vs_rfr_predicted[:10]
```

```python
# Evaluate the model
mse = mean_squared_error(y_train, mlp_predictions)
rmse = mse ** 0.5
r2 = r2_score(y_train, mlp_predictions)

print(f"Mean Squared Error: {mse}")
print(f"Root Mean Squared Error: {rmse}")
print(f"R-squared: {r2}")
```
