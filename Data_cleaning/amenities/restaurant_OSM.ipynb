```python
import requests

def reverse_geocode(lat,lng,api_key):
    geocode_url= 'https://maps.googleapis.com/maps/api/geocode/json'

    params= {'latlng':f'{lat},{lng}', 'key': api_key}

    response=requests.get(geocode_url,params=params)
    outcome=response.json()

    if outcome['status'] =='OK':
        final_outcome=outcome['results'][0]['formatted_address']
        return final_outcome
    else:
        raise Exception(f"Error in geocoding request: {outcome['status']}")
```

```python
for index,row in restaurants_gdf_details.iterrows():
    if row['restaurant_name']=='unknown':
        try:
            lat=row['centroid'].y
            lng=row['centroid'].x
            name= reverse_geocode(lat,lng,api_key)
            restaurants_gdf_details.at[index, 'restaurant_name'] =name
            print(f'Success geocoding location {row['restaurant_name']}')
        except Exception as e:
            print(f'Error geocoding address at {row['restaurant_name']} :{e}')
```
